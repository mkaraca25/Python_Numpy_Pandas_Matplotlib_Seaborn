```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x=np.linspace(0,10,100)
plt.plot(x,np.sin(x),"--")
plt.plot(x,np.cos(x),"-")
plt.xlabel("X ekseni")
plt.ylabel("Y ekseni")
plt.legend(("sin","cos"))
```

```python
fig=plt.figure()

plt.subplot(2,1,1)
plt.plot(x,np.sin(x),"r-",linewidth=4)
plt.xlabel("X ekseni")
plt.ylabel("Y ekseni")
plt.legend(("sin","cos"))
plt.subplot(2,1,2)
plt.plot(x,np.cos(x),"--",linewidth=2)
plt.xlabel("X ekseni")
plt.ylabel("Y ekseni")
plt.legend(("sin","cos"))
```

```python
fig.savefig("data.png")#kaydetme
```

```python
import seaborn as sns
import pandas as pd
iris=sns.load_dataset("iris")
iris.head()
```

```python
iris["petal_length"].plot()#Cizgi grafigi
```

```python
sns.lineplot(data=iris,y="petal_length",x=iris.index,hue="species")
```

```python
fig,ax=plt.subplots(figsize=(5,5))
ax.scatter(iris["sepal_length"],iris["sepal_width"])
ax.set_title("Iris Dataset")
ax.set_xlabel("sepal_length")
ax.set_ylabel("sepal_width")
```

```python
iris.groupby(by="species").mean()
```

```python
iris.groupby(by="species").max()
```

```python
iris.groupby(by="species").min()
```

```python
iris.groupby(by="species").sum()
```

```python
iris.groupby(by="species").mean()["sepal_length"]
```

```python
data=iris.groupby(by="species").max()["sepal_length"]
ax=sns.barplot(x=data.index,y=data.values)
ax.set_title("Sepal length means")
ax.set_ylabel("Max")
```

```python
sns.histplot(data=iris,x="sepal_width",bins=30)
```

```python
sns.histplot(data=iris,x="sepal_width",hue="species")
```

```python
iris.corr()
```

```python
sns.heatmap(iris.corr(),annot=True)
```

```python
sns.boxplot(data=iris,y="sepal_length",x="species")
```

```python
ax=plt.subplots(1,1,figsize=(10,8))
iris['species'].value_counts().plot.pie(explode=[0.1,0.1,0.1],autopct='%1.1f%%',shadow=True,figsize=(10,8))
plt.title("Iris Species %")
plt.show()
```
