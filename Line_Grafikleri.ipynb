## Line Grafikleri

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
%matplotlib inline
plt.style.use("seaborn-whitegrid")
fig=plt.figure()
ax=plt.axes()
x=np.linspace(0,10,100)
y=np.sin(x)
ax.plot(x,y)
```

```python
plt.plot(x,np.cos(x),color="k")
plt.plot(x,np.sin(x-1),color="r")
```

```python
plt.plot(x,x+1,linestyle="solid") #- ,-r
plt.plot(x,x+3,linestyle="dashed")#-- , --b
plt.plot(x,x+5,linestyle="dashdot")#-. ,-.g
plt.plot(x,x+7,linestyle="dotted")#: , :k
```

```python
plt.plot(x,np.sin(x))
plt.xlim(-5,20)
plt.ylim(-3,3)
```

```python
plt.plot(x,np.sin(x))
plt.axis([-5,20,-3,3])
```

```python
plt.plot(x,np.sin(x))
plt.axis("tight")
plt.title("Bir sinus egrisi")
plt.ylabel("sin(x degerleri)")
plt.xlabel("X degerleri")
```

```python
plt.plot(x,x+2,"-r",label="Cizgili")
plt.plot(x,x+4,":b",label="Noktali")
plt.legend()
```
