## Pandas and Seaborn ile pratik grafikler

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
iris=pd.read_csv("iris.txt",sep="\t")
iris.head()
```

```python
iris.Canak_yaprak_boyu.plot()
```

```python
iris.plot()
```

```python
fig,axes=plt.subplots(1,2)
veri=pd.Series(np.random.randn(10),index=list("ABCDEFGHIJ"))
veri.plot.bar(ax=axes[0],color="r",alpha=1)
veri.plot.barh(ax=axes[1],color="b",alpha=1)
```

```python
df=pd.DataFrame(abs(np.random.randn(5,3)),
                index=["bir","iki","uc","dort","bes"],columns=pd.Index(list("ABC")))
df
```

```python
df.plot.bar()
```

```python
df.plot.barh(stacked=True,alpha=1)
```

```python
bahsis=pd.read_csv("tips.csv")
bahsis.head()
parti_say=pd.crosstab(bahsis["day"],bahsis["size"])
parti_say
```

```python
parti_say.plot.bar()
```

```python
import seaborn as sns
bahsis["bahsis_yuzde"]=bahsis["tip"]/(bahsis["total_bill"]-bahsis["tip"])
bahsis.head()
```

```python
sns.barplot(x="day",y="bahsis_yuzde",hue="time",data=bahsis,orient="v")
```

```python
bahsis["bahsis_yuzde"].plot.hist(bins=70)
```

```python
sns.distplot(bahsis["bahsis_yuzde"],bins=50,color="r")
```

```python
sns.regplot("Canak_yaprak_boyu","Tac_yaprak_eni",data=iris)
```

```python
sns.pairplot(iris,hue="Turu")
```

```python
bahsis.head()
```

```python
sns.catplot(x="day",y="bahsis_yuzde",hue="time",col="smoker",kind="bar",data=bahsis)
```

```python
sns.catplot(x="day",y="bahsis_yuzde",row="time",col="smoker",kind="bar",data=bahsis)
```
