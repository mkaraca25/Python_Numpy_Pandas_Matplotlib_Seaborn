```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
plt.style.use("fivethirtyeight")
```

```python
veri=pd.Series(np.random.randn(1000).cumsum())
veri.plot()
```

```python
df=pd.DataFrame(np.random.randn(100,4),columns=list("ABCD"))
df=df.cumsum()
df.plot()
```

```python
df.iloc[10].plot(kind="bar")
```

```python
df.iloc[10].plot.bar()
```

```python
vs=pd.DataFrame(np.random.randn(7,3),columns=list("ABC"))
vs.plot.bar()
```

```python
vs.plot.bar(stacked=True)
```

```python
iris=pd.read_csv("iris.txt",sep="\t")
iris.head()
iris.plot.hist(alpha=1,stacked=True,bins=25)
```
