## Scatter Plot

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
%matplotlib inline
plt.style.use("seaborn-whitegrid")
```

```python
iris=pd.read_csv("iris.txt", sep="\t")
iris.head()
```

```python
fig=plt.figure()
ax=plt.axes()
plt.plot(iris.Canak_yaprak_boyu,iris.Canak_yaprak_eni, "rv",markersize=10)
```

```python
plt.scatter(iris.Canak_yaprak_boyu,iris.Canak_yaprak_eni,s=75,c="m",
           marker="v",edgecolor="black",linewidths=1,alpha=1)
```

```python
plt.style.use("seaborn-whitegrid")
plt.xkcd()
renk=np.random.randint(0,10,150)
boyut=np.random.randint(50,200,150)
plt.scatter(iris.Canak_yaprak_boyu,iris.Canak_yaprak_eni,s=boyut,c=renk,cmap="viridis",alpha=1)
cbar=plt.colorbar()
cbar.set_label("Seviyeler")
plt.xlabel("sepal en")
plt.ylabel("petal en")
plt.title("Iris cicegin sepal en & petal en")
```
